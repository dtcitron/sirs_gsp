### Generate phase diagrams

```python
alphas = [.1, .3, 1, 3, 10];
r0s = [.5, .9, 1, 2, 5];
```

```python
ad = sirs_diagram_endemic(100, r0s, alphas, g = 1, maxtime = 25, seed = 1, nruns = 100, fname = None)
```

```python
adict = sirs_diagram(1000, r0s, alphas, g = 1, maxtime = 10, dt = .1, nruns = 10, fname = None)
```

```python
# Phase diagram dictionary of average trajectories, 
# Infecteds vs. time for each parameter combination
mdict = {}
for r0 in r0s:
    for alpha in alphas:
        tg, Xg, Yg = adict[alpha, r0]
        tavg, Savg, Sstd, Iavg, Istd = grid_avg(tg, Xg, Yg)
        mdict[alpha, r0] = Iavg
mdict
```
