### Demonstrate generation of single trajectories

```python
# SI
[t,X,Y] = si_gsp(n=100, r0=1.2, ii=1, tmax = 10, seed = 0)
```

```python
si_traj = np.array([[ti, X[ti]] for ti in t]).transpose()
plt.plot(si_traj[0], si_traj[1])
ylim(0,100)
xlabel('Time')
ylabel('Susceptibles');
```

```python
# SIS
[t,X,Y] = sis_gsp(n=100, r0=3, g = 1, ii=10, tmax = 10, seed = 1)
```

```python
sis_traj = np.array([[ti, Y[ti]] for ti in t]).transpose()
plt.plot(sis_traj[0], sis_traj[1])
xlabel('Time')
ylabel('Infecteds')
ylim(0,100);
xlim(0,10);
```
