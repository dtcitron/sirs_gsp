### Demonstrate generation of single trajectories

```python
# SI
[t,X,Y] = si_gsp(n=100, r0=1.2, ii=1, tmax = 10, seed = 0)
```

```python
si_traj = np.array([[ti, X[ti]] for ti in t]).transpose()
plt.plot(si_traj[0], si_traj[1])
ylim(0,100)
xlabel('Time')
ylabel('Susceptibles');
```

```python
sis_traj = np.array([[ti, Y[ti]] for ti in t]).transpose()
plt.plot(sis_traj[0], sis_traj[1])
xlabel('Time')
ylabel('Infecteds')
ylim(0,100);
xlim(0,10);
```

### Generate a group of trajectories, coarse-grain the group in time and plot together in parallel

```python
# SI trajectories
t, X = si_group(n = 100, r0 = 1.2, tmax = 10, seed = 0, nruns = 10)
tg, Xg = gsp_trajectory_grid(t, X)
```

```python
figure()
for i in range(len(tg)):
    plt.plot(tg[i], Xg[i])
title("SI trajectories")
xlim(0, 10)
ylim(0,100);
```

```python
#SIS trajectories
t, X= sis_group(n = 1000, r0 = 1.2, g = 1, ii = 500, tmax = 10, seed = 0, nruns = 10)
tg, Xg = gsp_trajectory_grid(t, X)
```

```python
figure()
for i in range(len(tg)):
    plt.plot(tg[i], Xg[i])
title("SIS trajectories")
xlim(0, 10)
ylim(0, 1000);
```

```python
# SIRS trajectories
t, X, Y = sirs_group(n = 500, r0 = 2, g = 1, rho = 1, ii = 100, tmax = 10, seed = 0, nruns = 10)
tg, Xg, Yg = gsp_trajectory_grid(t, X, Y)
```

```python
figure()
for i in range(len(tg)):
    plt.plot(tg[i], Yg[i])
title("SIRS trajectories")
xlim(0, 10)
ylim(0, 500);
```
