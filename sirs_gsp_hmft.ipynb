# Examine the case of annealed graphs with nontrivial degree distributions

```python
def theta(pks, ks, kmean, ik):
    return np.dot(pks*ks, ik)/kmean

def iks(pks, ks, kmean, r0, rho, thet):
    b = r0/kmean
    return b*ks*thet/(1 + (1+1./rho)*b*ks*thet)
    
def find_fp(pks, ks, kmean, r0, rho, tol = 1e-9):
    ik_init = 0.
    ik_final = (1-1./r0)*pks
    maxcount = 100000000
    c = 0
    while np.sum(np.abs(ik_init-ik_final)) > tol and c < maxcount:
        ik_init = ik_final
        thet = theta(pks, ks, kmean, ik_final)
        ik_final = iks(pks, ks, kmean, r0, rho, thet)
        c += 1
    return ik_final

def ex_traj(t, X, Y):
    K = len(X[t[0]])
    ttrajs = np.array(t)
    Xtrajs = np.array([[X[i][j] for i in t] for j in range(K)])
    Ytrajs = np.array([[Y[i][j] for i in t] for j in range(K)])
    return ttrajs, Xtrajs, Ytrajs
```

```python
# parameters
n = 100
r0 = 1.5
g = 1
rho = 87
ii = n//10
tmax = 50
seed = 20
```

```python
# Compare to an annealed random regular graph
pk = {4:1.}
ks = np.array(sorted(pk.keys()))
pks = np.array([pk[k] for k in ks])
kmean = np.dot(ks, pks)
```

```python
# these are the results for fully mixed graphs
# all three quantities should be equal at the bottom
ik = find_fp(pks, ks, kmean, r0, rho, tol = 1e-9)
ik, theta(pks, ks, kmean, ik), rho/(1.+rho)*(1 - 1./r0)
```
